```python
class Node(object):
    def __init__(self, v=None):
        self.data = v
        self.parent = None
        self.left = None
        self.right = None
```

```python
class BinaryTree(object):
    def __init__(self):
        self.root = None
    
    def insert(self, data):
        new_node = Node(data)
        if self.root is None:
            self.root = new_node
        else:
            curr = self.root
            while curr is not None:
                p_node = curr
                if curr.data > data:
                    curr = curr.left
                else:
                    curr = curr.right
            
            new_node.parent = p_node
            if p_node.data > data:
                p_node.left = new_node
            else:
                p_node.right = new_node
    
    def find(self, data):
        res = []
        curr = self.root

        while curr is not None:
            p_node = curr
            if curr.data > data:
                curr = curr.left
            else:
                if curr.data == data:
                    res.append(curr)
                curr = curr.right

        return res

    def delete(self, data):
        rounds = len(self.find(data))
        for _ in range(rounds):
            self._delete(data)
    
    def _delete(self, data):
        p = self.root
        pp = None

        while p is not None and p.data != data:
            pp = p
            if data < p.data:
                p = p.left
            else:
                p = p.right
        
        if p is None:
            return
        
        if p.left is not None and p.right is not None:
            minimum_p = p.right
            minimum_pp = p
            while minimum_p.left is not None:
                minimum_pp = minimum_p
                minimum_p = minimum_p.left
            p.data = minimum_p.data
            p = minimum_p
            pp = minimum_pp
        
        if p.left is not None:
            child = p.left
        elif p.right is not None:
            child = p.right
        else:
            child = None
        
        if pp is None:
            self.root = child
        elif pp.left == p:
            pp.left = child
        else:
            pp.right = child

    def get_height(self):
        return self._get_height(self.root)
    
    def _get_height(self, node):
        if node is None:
            return 0
        current_height = max(self._get_height(node.left),
                             self._get_height(node.right))
        return current_height + 1
            
    def in_order(self):
        if self.root is None:
            return []
        
        return self._in_order(self.root)

    def _in_order(self, node):
        if node is None:
            return []
        
        ret = []
        ret.extend(self._in_order(node.left))
        ret.append(node.data)
        ret.extend(self._in_order(node.right))
        return ret
    
    def __repr__(self):
        return str(self.in_order())
```

```python
class Node:
    def __init__(self, name):
        self.children = []
        self.name = name

    def addChild(self, name):
        self.children.append(Node(name))
        return self

    def depthFirstSearch(self, array):
        if self is None:
            return []
        for c in self.children:
            c.depthFirstSearch(array)
        array.extend(self.name)
        return array
    
    def depthFirstSearch2(self, array):
        array.append(self.name)
        for c in self.children:
            c.depthFirstSearch2(array)
        return array
    
    def __repr__(self):
        return str(self.name)
```
